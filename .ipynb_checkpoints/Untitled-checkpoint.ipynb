```python
class ParseCSV:
    def __init__ (self,nama_file):
        self.nama_file = nama_file      #membuat variable self.nama_file berisi file dari name_file 
        self.df = self.read_file()
    def read_file(self):
        try:
            self.df = pd.read_csv(self.nama_file)
            return self.df
        except:
            self.df=1
            return self.df
    
    def display_file(self):
        return(self.df)
    def nilai_atas(self):
        print(self.df.head())
    def nilai_bawah(self):
        print(self.df.tail())
    def informasi(self):
        print(self.df.info())
    def hist_DATE(self):
        #print(self.df['Date'])
        self.df['Date'] = pd.to_datetime(self.df['Date'],infer_datetime_format=True)
        plt.clf()
        self.df['Date'].map(lambda d: d.month).plot(kind='hist')
        plt.title("Date")
        plt.show()
    def hist_AAPL(self):
        #return (self.df['AAPL'])
        num_bins = 6
        n, bins, patches = plt.hist(self.df['AAPL'], num_bins)
        plt.title("AAPL")
        plt.show()
    def hist_IBM(self):
        #return (self.df['AAPL'])
        num_bins = 6
        n, bins, patches = plt.hist(self.df['IBM'], num_bins)
        plt.title("IBM")
        plt.show()
    def hist_CSCO(self):
        #return (self.df['AAPL'])
        num_bins = 6
        n, bins, patches = plt.hist(self.df['CSCO'], num_bins)
        plt.title("CSCO")
        plt.show()
    def hist_MSFT(self):
        #return (self.df['AAPL'])
        num_bins = 6
        n, bins, patches = plt.hist(self.df['MSFT'], num_bins)
        plt.title("MSFT")
        plt.show()
    def plt_AAPL(self):
        self.df['Date'] = pd.to_datetime(self.df['Date'],infer_datetime_format=True)
        x_data = self.df['Date']
        y_data = self.df['AAPL']
        plt.scatter(x_data, y_data) # scatter plot
        plt.title("Scatter AAPL")
        plt.show()
        #plt.plot(x, y, 'o', color='black');
    def plt_IBM(self):
        self.df['Date'] = pd.to_datetime(self.df['Date'],infer_datetime_format=True)
        x_data = self.df['Date']
        y_data = self.df['IBM']
        plt.scatter(x_data, y_data) # scatter plot
        plt.title("Scatter IBM")
        plt.show()
    def plt_CSCO(self):
        self.df['Date'] = pd.to_datetime(self.df['Date'],infer_datetime_format=True)
        x_data = self.df['Date']
        y_data = self.df['CSCO']
        plt.scatter(x_data, y_data) # scatter plot
        plt.title("Scatter CSCO")
        plt.show()
    def plt_MSFT(self):
        self.df['Date'] = pd.to_datetime(self.df['Date'],infer_datetime_format=True)
        x_data = self.df['Date']
        y_data = self.df['MSFT']
        plt.scatter(x_data, y_data) # scatter plot
        plt.title("Scatter MSFT")
        plt.show()
def main():
    x = ParseCSV("stocks.csv")
    print(x.display_file())
    
    #print('----------------------------------')
    #print('         Nilai 5 teratas          ')
    #print('----------------------------------')
    #x.nilai_atas()
    #rint('----------------------------------')
    #rint('         Nilai 5 terbawah         ')
    #print('----------------------------------')
    #x.nilai_bawah()
    #print('----------------------------------')
    #print('            Informasi             ')
    #print('----------------------------------')
    x.hist_DATE()
    x.hist_AAPL()
    x.hist_IBM()
    x.hist_CSCO()
    x.hist_MSFT()
    x.plt_AAPL()
    x.plt_IBM()
    x.plt_CSCO()
    x.plt_MSFT()
    
```

```python
main()
```
